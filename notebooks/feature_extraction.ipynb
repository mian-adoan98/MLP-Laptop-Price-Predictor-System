Those values will be extracted and considered as binary features:

+ Bluetooth

```python
# Instantiate feature extractor object
feature_extractor = FeatureExtractor(dataset, "Features")
feature_df = pd.DataFrame()

# Extract features values from Feature
binary_feature_values = ["Touchscreen", "Bluetooth", "Wi-Fi", "Webcam(Built-In)", "Microphone(Built-In)", "Display"]

feature_df["Touchscreen"] = feature_extractor.extract("Touchscreen")
feature_df["Bluetooth"] = feature_extractor.extract("Bluetooth")
feature_df["Webcam(Built-In)"] = feature_extractor.extract("Webcam")
feature_df["Microphone(Built-In)"] = feature_extractor.extract("Microphone")
feature_df["Display"] = feature_extractor.extract("Display")
```
